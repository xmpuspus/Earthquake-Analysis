```
import pandas as pd
```

```
data = pd.read_csv('query (1).csv')
```

```
data.head()
```

```
data['datetime'] = pd.to_datetime(data['time'])
```

```
data['year'] = data['datetime'].dt.year
data['month'] = data['datetime'].dt.month
```

```
data['Country'] = data['place'].str.split(',').apply(lambda x: x[-1].strip())
```

```
data['Country'].value_counts().head(20)
```

```
data.shape
```

```
data.dtypes
```

```
data = data.sort_values('datetime', ascending=True)
data['eq_count'] = data.groupby('Country').cumcount()
```

```
data.head()
```

```
crosstab = pd.crosstab(data['Country'], [data['year']], values=data['eq_count'], aggfunc=max)
```

```
# Reformat data to fillforward for empty years
crosstab_new = crosstab.T.fillna(method='ffill').fillna(0).T
```

```
crosstab_new.head()
```

```
crosstab_new.reset_index().to_csv('eq_pivot.csv', index=False)
```
